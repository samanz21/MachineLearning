## Linear Regression Concept

<p>Linear Regression is used to predict the value of a variable based on the value of another variable


Linear regression is a fundamental statistical method used for modeling the relationship between a dependent variable (y) and one or more independent variables (x). It's one of the simplest and most commonly used techniques in machine learning and statistics for predictive analysis. Linear regression is advantageous due to its simplicity and ease of interpretation.
<br>

The goal of linear regression is to find the best-fitting line (or hyperplane) that minimizes the sum of squared residuals (SSR), often done using the method of ordinary least squares.

- It assumes a linear relationship between the variables, expressed as y = Œ≤‚ÇÄ + Œ≤‚ÇÅx‚ÇÅ + ‚ãØ + Œ≤·µ£x·µ£ + Œµ, where Œ≤‚ÇÄ, Œ≤‚ÇÅ, ..., Œ≤·µ£ are regression coefficients and Œµ is the random error term.
- The predicted response, ∆í(x), aims to be as close as possible to the actual response, y, for each observation.
- Coefficient of determination, R¬≤, measures the proportion of the variance in the dependent variable that is predictable from the independent variables. It ranges from 0 to 1, where higher values indicate a better fit.
- R¬≤ = 1 indicates a perfect fit, meaning the predicted values exactly match the observed values.


#### The basic concept is:  <br>
 - linear regression assumes a linear relationship between the independent variable(s) and the dependent variable. Mathematically, it can be represented as:

y = ùëè‚ÇÄ + ùëè‚ÇÅX + Œµ, where:

- y is the dependent variable (the variable we want to predict).
- ùëè‚ÇÄ is the intercept, representing the value of y when x is zero.
- ùëè‚ÇÅ is the slope of the regression line, representing the change in y for a one-unit change in x.
- X is the independent variable (the variable used to make predictions).
- Œµ is the error term, representing the difference between the predicted value and the actual value of y.

## Simple Linear Regression



Simple linear regression predicts a single output variable based on one input variable.

- aims to find the best-fitting line (regression line) through a set of data points by minimizing the sum of squared residuals (SSR).
- The intercept (ùëè‚ÇÄ) represents the point where the regression line crosses the y-axis, while the slope (ùëè‚ÇÅ) determines its steepness.
- Residuals are the differences between actual and predicted values.

The goal is to minimize these differences, making the predicted values as close as possible to the observed data points .


### Example with scikit-learn


Import libraries and create dummy data for a simple linear regression problem.

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

X = np.array([2, 4, 7, 10, 14, 18]).reshape((-1, 1))
y = np.array([4, 7, 9, 12, 16, 24]).reshape((-1, 1))
```

Plotting Original Data(dummy data)

```python
plt.scatter(X, y, color='blue', label='Original data')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

 The model is trained on the input data X and corresponding target values y using the fit method. The coefficient of determination R-squared ($R^2$) is calculated using the score method, and predictions are made using the predict method on the input data X.

```python
model = LinearRegression()
model.fit(X, y)
r_sq = model.score(X, y)
y_pred = model.predict(X)
```

The predictions are made using a Simple Linear Regression model previously trained. The equation for predicting y values based on X is manually computed using the model's intercept and coefficient, then new input data X_new is created and reshaped to a column vector. Predictions for the new data are generated using the model's predict method.

```python
y_pred_eq = model.intercept_ + model.coef_ * X
X_new = np.array([3, 6, 9, 12, 15, 16.5, 19]).reshape((-1, 1))
y_new_pred = model.predict(X_new)
```

Plotting simple linear regression

```python
print(f"coefficient of determination: {r_sq}")
print(f"intercept: {model.intercept_}")
print(f"slope: {model.coef_}")


plt.xlim(0, 20)
plt.ylim(0, 25)
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(X, y, color = "blue", label = "Original Data")
plt.scatter(X, y_pred, color = "red", label = "Predicted Points")
plt.scatter(X_new, y_new_pred, color = "cyan", label = "New Predicted Points")

line_X_ends = np.array([[0], [20]])
line_y_ends = model.predict(line_X_ends)
plt.plot(line_X_ends, line_y_ends, color='green', label='Linear regression')
plt.title(f"Simple Linear Regression\n$y$ = {model.intercept_} + {model.coef_[0]} $x$\n$R^2$ = {r_sq:.2f}")

plt.legend()
plt.show()
```

## Multiple Linear Regression


Multiple Linear Regression involves the use of two or more independent variables to generate an estimated regression function. It extends the concept of simple linear regression, where only one independent variable is used, to predict a value based on multiple variables simultaneously. By incorporating multiple predictors, the model aims to capture more complex relationships and provide a more accurate prediction of the dependent variable.

- Y = ùëè‚ÇÄ + ùëè‚ÇÅùë•‚ÇÅ + ùëè‚ÇÇùë•‚ÇÇ+ ‚Ä¶‚Ä¶ ùëènùë•n
- Y = Dependent variable and x1, x2, x3, ‚Ä¶‚Ä¶ xn = multiple independent variables
- function represents a regression plane in three-dimensional space.

The objective is to find weights (ùëè‚ÇÄ, ùëè‚ÇÅ, ùëè‚ÇÇ) that minimize the sum of squared residuals (SSR). With more than two independent variables, the function extends to ùëì(ùë•‚ÇÅ, ‚Ä¶, ùë•·µ£) = ùëè‚ÇÄ + ùëè‚ÇÅùë•‚ÇÅ + ‚ãØ +ùëè·µ£ùë•·µ£, where ùëü is the number of inputs, yielding ùëü + 1 weights to determine.

 Important note - Multiple Linear Regression makes several assumptions, such as linearity, independence of errors, constant variance of errors, and normality of errors, which should be checked before interpreting the results of the regression analysis.

The data set below contains some information about cars.

```python
import pandas as pd
# change the path according to csv file location in gGoogle Drive
# df = pd.read_csv('/content/drive/MyDrive/machine learning /assignment /data.csv')
df = pd.read_csv('data.csv') # if in same folder
df
```


Import Pandas and sklearn module. Extract the data from csv file(data.csv) the make a list of the independent values and call this variable X. Put the dependent values in a variable called y

```python
import pandas as pd
from sklearn import linear_model
import numpy as np
import plotly.graph_objects as go

# Load data
# df = pd.read_csv('/content/drive/MyDrive/machine learning /assignment /data.csv')
df = pd.read_csv('data.csv') 

# Extract features and target
X = df[['Weight', 'Volume']]
y = df['CO2']
```

To generate a linear regression object, we will utilize the LinearRegression() method from the sklearn, module. The fit() method is an object method that fills the regression object with data describing the relationship by accepting the independent and dependent values as parameters

After train the model, now we have a regression object that are ready to predict CO2 values based on a car's weight and volume

```python
# Train the model
regr = linear_model.LinearRegression()
regr.fit(X, y)

# Predict CO2 emission for all data points
predictedCO2 = regr.predict(X)
df['Predicted_CO2'] = predictedCO2


r_sq = regr.score(X, y)
coef0 = regr.intercept_
coef1 = regr.coef_.squeeze()[0]
coef2 = regr.coef_.squeeze()[1]
```


Plot the data to enhance understanding with visualization. In a 3D scatter plot, one can see the original data points, predicted data points, and regression plane more clearly.

```python
# Plot the data
fig = go.Figure()

# Add the original data points
fig.add_trace(go.Scatter3d(x=X['Weight'], y=X['Volume'], z=y, mode='markers',
                           marker=dict(size=5), name='Data'))

# Add the predicted data points
fig.add_trace(go.Scatter3d(x=X['Weight'], y=X['Volume'], z=predictedCO2, mode='markers',
                           marker=dict(size=5, color='red'), name='Predicted'))

# Generate points for the regression plane
mesh_weight, mesh_volume = np.meshgrid(np.linspace(X['Weight'].min(), X['Weight'].max(), 10),
                                       np.linspace(X['Volume'].min(), X['Volume'].max(), 10))
mesh_CO2 = regr.predict(np.array([mesh_weight.ravel(), mesh_volume.ravel()]).T).reshape(mesh_weight.shape)

# Add the regression plane
fig.add_trace(go.Surface(x=mesh_weight, y=mesh_volume, z=mesh_CO2, opacity=0.8, colorscale='Viridis'))

# Set axis labels
fig.update_layout(scene=dict(xaxis_title='Weight', yaxis_title='Volume', zaxis_title='CO2 Emission'))

title_text = "y = (%0.3f) + (%0.3f) X1 + (%0.3f) X2 [R^2 = %0.3f]" % (coef0, coef1, coef2, r_sq)

# Set title
fig.update_layout(
                  width=720,
                  height=600,
                  scene_camera={"eye": {"x": -1, "y": -1.5, "z": 1.5}},
                  title={"text": title_text,
                         "font": {"size": 20},
                         "x": 0.5,
                         "xanchor": "center"
                         }
                  )

# Show plot
fig.show()
```

Using Multiple Linear Regression, we get the predictionof how many CO2 release for every kilometer. This prediction aids in understanding the potential environmental impact of the vehicle, aiding in decision-making regarding emissions reduction strategies

### Real Life Situation

1. Business Forecasting: <br>
Linear regression can be used in business to forecast sales, demand for products, or any other relevant metric. By analyzing historical data, factors influencing sales (such as advertising expenditure, seasonality, economic indicators) can be identified and incorporated into the regression model. This allows businesses to make informed decisions about production levels, inventory management, and resource allocation.

2. Financial Analysis: <br>
Linear regression is widely used in finance for various purposes, such as asset pricing, risk management, and portfolio optimization. For example, analysts can use linear regression to estimate the relationship between the returns of a particular stock and the returns of a market index (such as the S&P 500). This information helps investors assess the risk and potential return of their investments and make strategic investment decisions.

3. Healthcare Analytics: <br>
Linear regression is valuable in healthcare for analyzing medical data and predicting outcomes. For instance, it can be used to predict patient readmission rates based on factors like age, comorbidities, length of hospital stay, and post-discharge care. Hospitals and healthcare providers can leverage these predictions to allocate resources effectively, identify high-risk patients for targeted interventions, and improve overall patient care outcomes.
